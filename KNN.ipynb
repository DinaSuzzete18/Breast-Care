```python
from sklearn.preprocessing import LabelEncoder
df = pd.DataFrame(data)
# Membuat objek LabelEncoder
label_encoder = LabelEncoder()
df['diagnosis'] = label_encoder.fit_transform(df['diagnosis'])
df
```

```python
X_lama= df.drop(columns='diagnosis', axis=1)
Y_lama= df['diagnosis']
```

```python
print(df.groupby('diagnosis').size())
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=0)
X, Y = ros.fit_resample(X_lama, Y_lama)
print("Shape of X_resampled:", X.shape)
print("Shape of y_resampled:", Y.shape)
print("Class distribution of y_resampled:")
print(pd.Series(Y).value_counts())
```

```python
X.describe ()
```

# DATA SPLIT

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_scaled, y_train)
```

```python
prediksiKNN=knn.predict(X_test_scaled)
prediksiKNN
```

```python
print('Confusion Matrix:', (confusion_matrix(y_test, prediksiKNN))) #Membuat confusion matrix
print('Classification Report', (classification_report(y_test, prediksiKNN)))#Membuat classification report
print('Akurasi: ', (accuracy_score(y_test, prediksiKNN)))#Menghitung akurasi model SVM
```
